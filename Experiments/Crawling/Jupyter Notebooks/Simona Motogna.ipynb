```python
def IsDate(line):
    if line == "":
        return False
    for c in line:
        if c not in "1234567890 ":
            return False
    return True
```

```python
cdate = None
authors = ""
title = ""

pubs = []
for line in data.split('\n'):
    print(line)
    if IsDate(line):
        cdate = line
        continue
    
    if authors == "":
        authors = line
    else:
        title = line
    
    if authors != "" and title != "":
        pubs.append((authors, title, cdate))
        authors = ""
        title = ""
      
```
